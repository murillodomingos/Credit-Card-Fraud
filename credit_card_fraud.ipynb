# **Credit Card Fraud Prediction**

1. **Dataset Overview**
- The dataset contains anonymized features due to privacy concerns. A total of 30 features are provided, including the transaction time, transaction amount, and 28 principal components (V1 to V28) obtained through PCA (Principal Component Analysis).

2. **Columns:**
- Time: Seconds elapsed between the first transaction and the current transaction.
- V1 - V28: Features resulting from PCA transformation (dimensionality reduction) for anonymization.
- Amount: The amount of the transaction.
- Class: The target variable( 0: Non-fraudulent transaction | 1: Fraudulent transaction).

3. **Main Goal**
- The primary objective is to develop a predictive model that accurately identifies fraudulent transactions while minimizing false positives and false negatives. Due to the extreme class imbalance, strategies like oversampling, undersampling, or anomaly detection models may be required.

```python
# Correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Define correlation threshold
threshold = 0.5

# Find correlated columns
high_corr_cols = set()
for i in range(len(corr.columns)):
  for j in range(i):
    if abs(corr.iloc[i, j]) >= threshold:
      colname1 = corr.columns[i]
      colname2 = corr.columns[j]
      high_corr_cols.add(colname1)
      high_corr_cols.add(colname2)

# Plot
for col1 in high_corr_cols:
  for col2 in high_corr_cols:
    if col1 != col2:
      sns.scatterplot(data=df, x=col1, y=col2)
      plt.show()
```
